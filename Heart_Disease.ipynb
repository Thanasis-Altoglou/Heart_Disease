## Part 03: Prepare the data for training

```python
#seperate the features from the labels into 2 objects, X and y
X = df.drop('target',axis=1)
y = df['target']
```

```python
#perform a train test split, with the test size of 10% and random state of 101
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

```python
#scale the data with standard scaler and fit only to the training data
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Part 04: Building the models


### 1. Logistic Regression with Cross Validation

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
#create the model
log_model = LogisticRegressionCV()
```

```python
#fit the model
log_model.fit(X_train_scaled,y_train)
```

```python
#report back the search's optimal parameters, especially the C value
log_model.C_
```

```python
#check the model's parameters
log_model.get_params()
```

```python
#report back the model's coefficients
log_model.coef_
```

Let's create a visualization of the coefficients by using a barplot of their values.

```python
#create a series of the coefficients
coefs  = pd.Series(index = X.columns, data=log_model.coef_[0])
coefs
```

```python
coefs = coefs.sort_values()
plt.figure(figsize=(10,4),dpi=150)
sns.barplot(x=coefs.index,y=coefs.values)
```

It can be observed that the most important features are the chest pain, the slope and the thalach.

Model Evaluation using:
1. classification report
2. confusion matric
3. ROC curve.

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,roc_curve,plot_roc_curve
```

```python
#evaluate the model on the testing subset
y_preds = log_model.predict(X_test_scaled)
```

```python
print(classification_report(y_test,y_preds))
```

The logistic regression model has 84% accuracy.


```python
#confusion matrix
plot_confusion_matrix(log_model,X_test_scaled,y_test)
```

Logistic regression model misclassifies 5 patients out of 31. Specifically, it predicts that 3 patients have heart disease while in the reality are healthy and 2 patients that are healthy while the have heart disease.

```python
#plot the ROC CURVE
plot_roc_curve(log_model,X_test_scaled,y_test)
```

### K Nearest  Neighbors 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#create the model
knn = KNeighborsClassifier(n_neighbors = 2) 
```

```python
#fit the model
knn.fit(X_train_scaled,y_train)
```

```python
#evaluate the model
preds = knn.predict(X_test_scaled)
```

```python
print(classification_report(y_test,preds))
```

**Elbow Method for Choosing Reasonable K Values**

```python
test_error_rates = []


for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train) 
   
    y_pred_test = knn_model.predict(X_test_scaled)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
```

```python
plt.figure(figsize=(4,2),dpi=200)
plt.plot(range(1,30),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

According to the plot, the most optimal K value is between 7 and 8. Lets create again the knn model with k value equals to 7.

```python
#create the model again
knn_model = KNeighborsClassifier(n_neighbors = 7) 
```

```python
#fit the model
knn_model.fit(X_train_scaled,y_train)
```

```python
#evaluate the model
predictions = knn_model.predict(X_test_scaled)
```

```python
predictions
```

```python
#classification report
print(classification_report(y_test,predictions))
```

The final model with 7 number of neighbours has been improved significantly compared with the first model. Knn model can predict whether a patient has heart disease or not with 94% accuracy.

```python
#confusion matrix
plot_confusion_matrix(knn_model,X_test_scaled,y_test)
```

Knn model misclasified only 2 patients. It predicts that two of them are healthy while in the reality have heart disease.

```python
#plot the ROC CURVE
plot_roc_curve(knn_model,X_test_scaled,y_test)
```
